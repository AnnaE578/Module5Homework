```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st

import warnings
warnings.filterwarnings("ignore")

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
Pymaceuticals_data_complete = pd.merge(mouse_metadata, study_results, how="left", on=["Mouse ID"])


# Display the data table for preview
Pymaceuticals_data_complete.head()
```

```python
# Checking the number of mice.
len(Pymaceuticals_data_complete["Mouse ID"].unique())
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
Duplicated_mouse_id = Pymaceuticals_data_complete.loc[Pymaceuticals_data_complete.duplicated(subset = ["Mouse ID", "Timepoint"])]
Duplicated_mouse_id["Mouse ID"].unique()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_Pymaceuticals_data = Pymaceuticals_data_complete.loc[Pymaceuticals_data_complete["Mouse ID"] != "g989"]
cleaned_Pymaceuticals_data 
```

```python
# Checking the number of mice in the clean DataFrame.
len(cleaned_Pymaceuticals_data["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
mean_tumor_volume = cleaned_Pymaceuticals_data.groupby("Drug Regimen").mean()["Tumor Volume (mm3)"]
mean_tumor_volume

median_tumor_volume = cleaned_Pymaceuticals_data.groupby("Drug Regimen").median()["Tumor Volume (mm3)"]
median_tumor_volume

variance_tumor_volume = cleaned_Pymaceuticals_data.groupby("Drug Regimen").var()["Tumor Volume (mm3)"]
variance_tumor_volume

std_tumor_volume = cleaned_Pymaceuticals_data.groupby("Drug Regimen").std()["Tumor Volume (mm3)"]
std_tumor_volume

SEM_tumor_volume = cleaned_Pymaceuticals_data.groupby("Drug Regimen").sem()["Tumor Volume (mm3)"]
SEM_tumor_volume

Pymaceuticals_summary = pd.DataFrame({
    "Mean Tumor Volume": mean_tumor_volume, "Median Tumor Volume": median_tumor_volume, "Tumor Volume Variance": variance_tumor_volume, 
     "Tumor Volume Std. Dev.": std_tumor_volume, "Tumor Volume Std. Err.": SEM_tumor_volume})

Pymaceuticals_summary
```

```python
mean_tumor_volume = cleaned_Pymaceuticals_data.groupby("Drug Regimen").mean()["Tumor Volume (mm3)"]
mean_tumor_volume
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
tumor_volume_data = cleaned_Pymaceuticals_data.groupby("Drug Regimen").agg({"Tumor Volume (mm3)": 
                                                    ["mean", "median", "var", "std", "sem"]})
tumor_volume_data
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
tumor_max = cleaned_Pymaceuticals_data.groupby("Mouse ID")["Timepoint"].max()
tumor_max = tumor_max.reset_index()
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
tumor_merged = tumor_max.merge(cleaned_Pymaceuticals_data, on = ["Mouse ID", "Timepoint"], how = "left")
tumor_merged
```

```python
# Put treatments into a list for for loop (and later for plot labels)

drug_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []


# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for drug in drug_list:  
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_vol = tumor_merged.loc[tumor_merged["Drug Regimen"] == drug, "Tumor Volume (mm3)"]
    # add subset 
    tumor_vol_data.append(tumor_vol)
    # Determine outliers using upper and lower bounds
    quartile = tumor_vol.quantile([.25, .5, .75])
    lower = quartile[.25]
    upper = quartile[.75]
    iqr = upper - lower
    upper_b = upper + (1.5*iqr)
    lower_b = lower - (1.5*iqr)
    
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
marker_style = dict(linestyle=':', color='0.8', markersize=10,
                    markerfacecolor="tab:red", markeredgecolor="tab:blue")

fig1, ax1 = plt.subplots()
ax1.set_title("Outlier Results")
ax1.set_ylabel("Final Tumor Volume (mm3)")
ax1.boxplot(tumor_vol_data, labels = drug_list, flierprops = marker_style)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomuli

capomulin_tumor_vol = cleaned_Pymaceuticals_data.loc[cleaned_Pymaceuticals_data["Mouse ID"] == "s185"]
capomulin_tumor_vol

plt.plot(capomulin_tumor_vol["Timepoint"], capomulin_tumor_vol["Tumor Volume (mm3)"])
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Capomulin Treatment of Mouse s185")
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
mouse_scatter_data = cleaned_Pymaceuticals_data.loc[cleaned_Pymaceuticals_data["Drug Regimen"] == "Capomulin"]
capomulin_regimen_ave = mouse_scatter_data.groupby("Mouse ID").mean()
capomulin_regimen_ave

plt.scatter(capomulin_regimen_ave["Weight (g)"],capomulin_regimen_ave["Tumor Volume (mm3)"])
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
weight = capomulin_regimen_ave.iloc[:,1]
tumor = capomulin_regimen_ave.iloc[:,3]
correlation = st.pearsonr(weight,tumor)
print(f"The correlation between both factors is {round(correlation[0],2)}")

x_values = capomulin_regimen_ave["Weight (g)"]
y_values = capomulin_regimen_ave["Tumor Volume (mm3)"]
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(5.8,0.8),fontsize=15,color="red")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```
